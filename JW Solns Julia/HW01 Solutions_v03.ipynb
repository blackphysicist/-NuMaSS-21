# Prob 2. Lines in the plane
a) Plot two lines on a single graph, with grid lines to read the point where they cross

```julia
# Method 1
line1(x)= 0.5x+1  # creates a "generic" function called line1
line2(x)=-x+4    # creates a "generic" function called line2
xmin,xmax= -5.0, 5.0
plot(line1,[xmin,xmax],label="y = 0.5x + 1")
# Now we use the plot! (called plot-bang or plot exclamation)
# so that we do not erase the previous plot
plot!(line2,[xmin,xmax],label="y = -x+4")
# We can then read off the intersectin point as (2,2)
```

```julia
# Suppose you wanted to see much closer around (2,2)? Initially, we were just guessing where the lines might cross.
# We simply change xmin and xmax and re-run the plots. We do not have to change the functions.
xmin,xmax= 1.9, 2.1 
plot(line1,[xmin,xmax],label="y = 0.5x + 1")
# Now we use the plot! (called plot-bang or plot exclamation)
# so that we do not erase the previous plot
plot!(line2,[xmin,xmax],label="y = -x+4")
```

```julia
# Method 2 (any plot is fine)
# Instead of letting x be a variable, we make define it as a vector of values of x, ranging from -10.0, to 10.0, 
# in increments of 0.1
x = -10.0:.1:10.0 # 201 values of x, count them!
nx=length(x)
println("number of elements of x = $nx")  # The dollar sign before nx tells the 
                                          # function println to print the VALUE of nx 
                                          # instead of the charcters nx
y= 0.5 * x .+ 1.0 #the "dot" before the "+" sign 
                   #tells Julia to add the 1.0 to each of the 201 
                   #values of 0.5 * x
 plot(x, y, label="y = 0.5x + 1") 
 y = -x .+ 4.
 # Now we use the plot! (called plot-bang or plot exclamation)
 # so that we do not erase the previous plot
 plot!(x, y, label="y = -x + 4")
 # Next we add some optional commands to make the plot a bit nicer
 # We use again plot-bang so that we apply these changes
 # to the current plot. Without the ! (bang), we erase the
 # plot and start a new one. 
 plot!(xaxis=(font(12), "X", -10:1:10), yaxis = (font(12), "Y", -6:1:10), leg=true, grid=true)
 plot!(fmt = :png)
```

```julia
# Method 3, Yet another way to plot
# now declare both of those equations as generic functions
# because x has been used, we'll use z as the dummy varaible
# z = -10.0:.1:10.0
f1(z) = 0.5z + 1.
f2(z) = -z + 4.

# Plot them on the same graph.  This shows another way to plot multiple things at once
plot(-10.0:10.0, [f1,f2])
plot!(xaxis = (font(12), "X", -10:1:10), yaxis = (font(12), "Y", -6:1:10), label=["y = 0.5x + 1", "y = -x + 4"], leg=true, grid=true)
```

The discriminat is negative thus the two roots are complex. There are no real roots, which is all we need to know for now.

```julia
# Plot the graph to see whether it crosses the x axis or not
# Define a function f that returns 2x^2 + 3x + 2
f(x::Float64) = 2 + 3 * x + 2 * x^2
plot(f, -5.0, 3, title="y = 2x^2 + 3x + 2")
plot!(xaxis=(font(12), "X", -6:1:3), yaxis=(font(12), "Y", 0:5:25), leg=false)
```
