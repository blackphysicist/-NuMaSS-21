# Pro2. Lines in the plane
a) Plot two lines on a single graph, with grids to read the cross points

```julia
 x = -10.0:.1:10.0 # 201 values of x, count them!
 y= 0.5 * x .+ 1.0 #the "dot" before the "+" sign 
                   #tells Julia to add the 1.0 to each of the 201 
                   #values of 0.5 * x
 plot(x, y, label="y = 0.5x + 1") 
 y = -x .+ 4.
 # Now we use the plot! (called plot-bang or plot exclamation)
 # so that we do not erase the previous plot
 plot!(x, y, label="y = -x + 4")
 # Next we add some optional commands to make the plot a bit nicer
 # We use again plot-bang so that we apply these changes
 # to the current plot. Without the ! (bang), we erase the
 # plot and start a new one. 
 plot!(xaxis=(font(12), "X", -10:1:10), yaxis = (font(12), "Y", -6:1:10), leg=true, grid=true)
 plot!(fmt = :png)
```

Below is a function used to calcualte intersection of two lines by linear algebra. It may out of the scope.

```julia
using LinearAlgebra
# This function is used to calculate the intersection point of two lines
function check_intersection(a1::Number, b1::Number, c1::Number, a2::Number, b2::Number, c2::Number)
    """ Check whehter there is intersections between two lines. If it has intersection, print the point;
        Otherwise, print no intersection.
        Input:
            a1, b1, c1: parameters of the line: a1x + b1y = c1
            a2, b2, c2: parameters of the line: a2x + b2y = c2
        Output:
            String: indicate whether ther is a intersection."""
    A = [a1 b1; a2 b2]
    # Check whether there is intersection
    determinant = det(A)
    # if determinant = 0, there is no intersection. 
    if determinant == 0
        return String("There is no intersection. Two lines are parallel!")
    else
        b = [c1; c2]
        intersection = inv(A) * b
        return String("There is intersection. The intersection is at $(intersection).")
    end
end

# Check if there is intersection point between y1 = 0.5*x .+ 1. and y2 = -x .+ 4.
check_intersection(0.5, -1, -1, -1, -1, -4)
    
    
```

```julia
# Define a function to claculate intersection of quadratic equation
function calculate_intersection_quad(a::Number, b::Number, c::Number, y::Number)
    """This function is used to claculate intersection of quadratic equation.
       Input:
            a, b, c, y: parameters of quadratic equation ax^2+bx+c = y, a is not equal to zero."""
    c = c - y
    delta = b^2 - 4 * a * c
    if delta > 0
        root1 = (-b + sqrt(delta)) / (2 * a)
        root2 = (-b - sqrt(delta)) / (2 * a)
        return String("There are two intersection points: $([root1, y]) and $([root2, y]).")
    elseif delta == 0
        root = -b / (2 * a)
        return String("There is one intersection point: $([root, y]).")
    else
        return String("There is no intersection point.")
    end
end

# Check if there is intersection point between f(x) = -4+x+2x^2 and y = 0
calculate_intersection_quad(2, 1, -4, 0)
```

Below is a function used to calculate discriminant.

```julia
# Creat a function to evaluate discriminant
function evaluate_discriminant(a::Number, b::Number, c::Number)
    """This function is used to evaluate discriminant of a quadratic equation
    Input:
        a, b, c: constants of a quadratic equation of form ax^2 + bx +c = 0, a is nonzero
    Output:
        The value of discriminant"""
    delta = b^2 - 4 * a * c
    return delta
end

# Test for quadratic equation: 2x^2 + 3x -2 = 0
@show discriminant = evaluate_discriminant(2, 3, -2)
```

```julia
# Calculate the roots
@show root1 = (-3 + sqrt(discriminant)) / (2 * 2)
@show root2 = (-3 - sqrt(discriminant)) / (2 * 2)
```

Below is a function used to calcualte the roots.

```julia
# Create a function to calculate the roots of the quadratic equation
function calculate_roots(a::Number, b::Number, c::Number)
     """This function is used to calculate the roots of a quadratic equation
    Input:
        a, b, c: constants of a quadratic equaiton of form ax^2 + bx +c = 0, a is nonzero
    Output:
        String: if there are real roots, show their value. Otherwise, print no real roots"""
    delta =  b^2 - 4 * a * c
    if delta > 0
        root1 = (-b + sqrt(delta)) / (2 * a)
        root2 = (-b - sqrt(delta)) / (2 * a)
        return String("Discriminant is greater than 0. There are two different real roots: $(root1) and $(root2).")
    elseif delta == 0
        root = -b / (2 * a)
        return String("Discriminant is equal to 0. There are two identical real roots: $(root).")
    else
        return String("Discriminant is negative. No real solutions. The roots are complex numbers.")
    end
end

# Test for quadratic equation: 2x^2 + 3x -2 = 0
roots = calculate_roots(2, 3, -2)
```

```julia
# Define a functiont to find the minimum value for a parabola that is curved upwards
function find_min(a::Number, b::Number, c::Number)
    min_x = -b / (2 * a)
    min_y = a * min_x^2 + b * min_x + c
    return min_x, min_y
end

# Plot minimum value of y = 2x^2 + 3x + 2
min_x, min_y = find_min(2, 3, 2)
plot(f, -5.0, 3, label="y = 2x^2 + 3x + 2")
plot!(xaxis=(font(12), "X", -6:0.5:3), yaxis=(font(12), "Y", 0:2.5:25), leg=true)
scatter!([min_x], [min_y], label="Minimum Value")
```
